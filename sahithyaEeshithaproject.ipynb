```python
import pandas as pd   # Import pandas library for data manipulation
import numpy as np    # Import numpy library for numerical operations
import matplotlib.pyplot as plt   # Import matplotlib for plotting graphs
from sklearn.linear_model import LinearRegression   # Import Linear Regression model
from sklearn.model_selection import train_test_split #Import function to split dataset into train and test
from sklearn.metrics import mean_squared_error, r2_score  # Import metrics to evaluate regression model
```

```python
df = pd.read_csv("Salary_data.csv")     # Load the CSV file into a pandas DataFrame
```

```python
print(df.head(5)) # Prints the first 5 rows of the DataFrame
```

```python
df = df[df["YearsExperience"] >= 1]  # Filter data to keep only rows where YearsExperience >= 1
```

```python
df["YearsExperience"] = df["YearsExperience"].astype(int)  # Convert YearsExperience column values to integers
```

```python
df.head(20)  #displays first 20 rows
```

```python
df   #displays full DataFrame
```

```python
plt.scatter(df["YearsExperience"], df["Salary"], color='purple')  # Creating scatter plot of YearsExperience vs Salary
plt.title("Years of Experience vs Salary") # Setting plot title  
plt.xlabel("Years of Experience") # Label x-axis
plt.ylabel("Salary")  # Label y-axis
plt.grid(False)  # Disable background grid
plt.show()    # Display the plot
```

```python
X = df[["YearsExperience"]] # Defining input features (independent variable) as DataFrame  # input feature
y = df["Salary"] # Define target variable (dependent variable) as Series # output label
```

```python
X  #display x values
```

```python
y   #display y values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) # Split data into training and testing sets (80% train, 20% test, random_state ensures reproducibility)
 
```

```python
X_train  #shows training input data
```

```python
X_test #shows testing input data
```

```python
model = LinearRegression()  # Create Linear Regression model object
model.fit(X_train, y_train)  # Train the model on training data (fit regression line)
```

```python
y_pred = model.predict(X_test)   # Predict target values (salary) for test input data
X_pred=model.predict(X_train)
```

```python
y_pred  #show predicted salaries for test data
```

```python
y_test
```

```python
X_test # display test input values again
```

```python
y_pred = model.predict(pd.DataFrame([[17]],
columns=['YearsExperience'])) # Predict salary for a new input: 17 years of experience
# (overwrites the earlier y_pred array with this new single prediction)
```

```python
y_pred #show salary prediction for 17 years experience
```

```python
plt.scatter(X_test, y_test, color='purple', label='Actual')# Scatter plot of actual test data points
X_test_sorted = np.sort(X_test.values.ravel())# Flatten test input values into 1D array and sort them in ascending order
X_test_sorted_df=pd.DataFrame(X_test_sorted,
columns=["YearsExperience"]) # Convert sorted array back into DataFrame with same column name
y_pred_sorted=model.predict(X_test_sorted_df) # Predict salary values for sorted test input data
plt.plot(X_test_sorted, y_pred_sorted, color='red', linewidth=2, label='Predicted') # Plot regression line using sorted test values (red line)
plt.title("Salary Prediction") # Set plot title
plt.xlabel("Years of Experience")    # Label x-axis
plt.ylabel("Salary")  # Label y-axis
plt.legend() # Show legend for Actual and Predicted points/line
plt.show()  # Display the plot
```
